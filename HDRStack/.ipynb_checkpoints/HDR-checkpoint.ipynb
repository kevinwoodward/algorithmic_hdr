```python
import numpy as np
import os
import cv2 as cv
import matplotlib.pyplot as plt
import math
from scipy.stats import linregress
import pickle
```

```python
with open("../WhiteImages/linearized_R.pickle") as file:
    linearized_R = pickle.load(file)
with open("../WhiteImages/linearized_G.pickle") as file:
    linearized_G = pickle.load(file)
with open("../WhiteImages/linearized_B.pickle") as file:
    linearized_B = pickle.load(file)
```

```python
# Images 27, 34, 37

# def readimg_RGB(img_path):
#     img = cv.imread(img_path)
#     return cv.cvtColor(img, cv.COLOR_BGR2RGB)



# img_a0 = readimg_RGB('IMG_0027.JPG') # Exposed to the right at T = 1/640
# img_a1 = readimg_RGB('IMG_0034.JPG') # Exposed at a1 * T = 1/125
# img_a2 = readimg_RGB('IMG_0037.JPG') # Exposed at a2 * T= 1/60

img_a0 = cv.imread('IMG_0027.JPG') # Exposed to the right at T = 1/640
img_a1 = cv.imread('IMG_0034.JPG') # Exposed at a1 * T = 1/125
img_a2 = cv.imread('IMG_0037.JPG') # Exposed at a2 * T= 1/60



img_list = [img_a0, img_a1, img_a2]
plt.imshow(img_a0)
plt.show()
plt.imshow(img_a1)
plt.show()
plt.imshow(img_a2)
plt.show()
```

```python
linearized_channels = []
for img in img_list:
    
    img_copy = img.copy()
    img_b_linearized, img_g_linearized, img_r_linearized = cv.split(img_copy)    

    b_g = (1./linearized_R.slope)
    g_g = (1./linearized_G.slope)
    r_g = (1./linearized_B.slope)

#     img_b_linearized = linearized_B.intercept*(img_b_linearized**b_g)
#     img_g_linearized = linearized_G.intercept*(img_g_linearized**g_g)
#     img_r_linearized = linearized_R.intercept*(img_r_linearized**r_g)

    img_b_linearized = np.power(img_b_linearized, b_g)
    img_g_linearized = np.power(img_g_linearized, g_g)
    img_r_linearized = np.power(img_r_linearized, r_g)

    linearized_channels.append((img_b_linearized.astype(np.float32), img_g_linearized.astype(np.float32), img_r_linearized.astype(np.float32)))
    
    
    

# plt.imshow(a0_r)
# plt.show()
# plt.imshow(a0_r_linearized)
# plt.show()
# plt.imshow(a0_g)
# plt.show()
# plt.imshow(a0_g_linearized)
# plt.show()
# plt.imshow(a0_b)
# plt.show()
# plt.imshow(a0_b_linearized)
# plt.show()
# img_linearized = a0_r_linearized + a0_g_linearized + a0_b_linearized
# plt.imshow(img_a0)
# plt.show()
# plt.imshow(img_linearized)
# plt.show()
```

```python
exposure_names = ["1/640", "1/125", "1/60"]
exposure_values = [1./640, 1./125, 1./60]
for img, exposure in zip(img_list, exposure_names):
    
    plt.imshow(img)
    plt.title("Exposure time: {0}th of a second".format(exposure))
    plt.show()
```

```python

def generateHistogram(channel, g_val, bin_count=256, range=(0,255)):
    bin_axis = np.zeros((bin_count,), dtype=int)
    value_list = channel.ravel()
    bin_width = math.pow(range[1], g_val)/bin_count
    bin_index = 0
    for value in value_list:
        bin_index = int((value / range[1]**g_val) * bin_count)
        bin_axis[bin_index] += 1
    return bin_axis
    
bin_axis = generateHistogram(linearized_channels[0][0], r_g)
#print((35. / 255**r_g) * 256)
```

```python
for linearized_img, exposure_name in zip(linearized_channels, exposure_names):
    r_channel = np.power(linearized_img[2], 1/r_g)
    plt.hist(r_channel.ravel(), bins=256, range=None, color='red')
    plt.title("Red channel histogram of {0}s exposure linearized".format(exposure_name))
    plt.xlabel("256 bins, range: {0, $255^{g_{red}}$}")
    plt.ylabel("Number of pixels")    
    plt.show()
    
    g_channel = np.power(linearized_img[1], 1/g_g)
    plt.hist(g_channel.ravel(), bins=256, range=None, color='green')
    plt.title("Green channel histogram of {0}s exposure linearized".format(exposure_name))
    plt.xlabel("256 bins, range: {0, $255^{g_{green}}$}")
    plt.ylabel("Number of pixels")
    plt.show()
    
    b_channel = np.power(linearized_img[0], 1/b_g)
    plt.hist(b_channel.ravel(), bins=256, range=None, color='blue')
    plt.title("Blue channel histogram of {0}s exposure linearized".format(exposure_name))
    plt.xlabel("256 bins, range: {0, $255^{g_{blue}}$}")
    plt.ylabel("Number of pixels")
    plt.show()
```

```python
a_values = (exposure_values[1]/exposure_values[0], exposure_values[2]/exposure_values[0])

trunc_exposure_names = list(exposure_names)
trunc_exposure_names.pop(0)

trunc_linearized = list(linearized_channels)
trunc_linearized.pop(0)

trunc_img_list = list(img_list)
trunc_img_list.pop(0)
    
print(np.array(linearized_channels).shape)

for linearized_img, a_value, exposure in zip(trunc_linearized, a_values, trunc_exposure_names):
    print(a_value)
    r_channel = np.float32(linearized_img[2]/a_value)
    plt.hist(r_channel.ravel(), bins=256, range=[0,math.pow(255, r_g)], color='red')
    plt.title("Red channel histogram of {0}s exposure linearized, divided by a = {1:.2f}".format(exposure, a_value))
    plt.xlabel("25 bins, range: {0, $255^{g_{red}}$}")
    plt.ylabel("Number of pixels")
    plt.show()
    
    g_channel = np.float32(linearized_img[1]/a_value)
    plt.hist(g_channel.ravel(), bins=256, range=[0,math.pow(255, g_g)], color='green')
    plt.title("Red channel histogram of {0}s exposure linearized, divided by a = {1:.2f}".format(exposure, a_value))
    plt.xlabel("25 bins, range: {0, $255^{g_{green}}$}")
    plt.ylabel("Number of pixels")
    plt.show()
    
    b_channel = np.float32(linearized_img[0]/a_value)
    plt.hist(b_channel.ravel(), bins=256, range=[0,math.pow(255, b_g)], color='blue')
    plt.title("Red channel histogram of {0}s exposure linearized, divided by a = {1:.2f}".format(exposure, a_value))
    plt.xlabel("25 bins, range: {0, $255^{g_{blue}}$}")
    plt.ylabel("Number of pixels")
    plt.show()
    
```

```python
hdr_lin_img = img_lin_a2.copy()

#copy over the values of a1
for x, row in enumerate(img_a2):
    for y, pixel in enumerate(row):
        #print(x,y)
        if 255 in pixel:
            hdr_lin_img[x][y] = img_lin_a1[x][y]
            
#copy over the values of a0
for x, row in enumerate(img_a1):
    for y, pixel in enumerate(row):
        #print(x,y)
        if 255 in pixel:
            hdr_lin_img[x][y] = img_lin_a0[x][y]
```

```python
hdr_b,hdr_g,hdr_r = cv.split(img_lin_a0)
print(hdr_b)
hdr_b = hdr_b/255**b_g
hdr_g = hdr_g/255**g_g
hdr_r = hdr_r/255**r_g
out = cv.merge((hdr_b, hdr_g, hdr_r))


plt.imshow(out)
```
